```python
import math


# Inputs:
# x0  : reference point x_0
# u0  : u(x) at x0
# u1  :  u(x) at x0 + h
# b2  : parameter from the cost function
# b3  : parameter from the cost function
# c,d : bounds of p domain [c,d]

def interval(x0, u0, u1, b2, b3, c, d):

    # difference appearing in the inequality

    diff = u0 - u1

    # Case 1.1:
    # If u0 - u1 < 0, the inequality
    # sqrt() <= u0 - u1
    # square root is nonnegative.
    if diff < 0:
        print("No possible p value (u0 - u1 < 0)")
        return None

    # Rearranging the inequality gives
    # (x0 - p)^2 >= ((u0 - u1)^2 - b2) / b3
    # The right-hand side is called rhs.
    rhs = (diff**2 - b2) / b3

    # If rhs < 0 then the inequality is satisfied
    # for every p in the domain [c,d].
    if rhs < 0:
        print("All p in [c,d] satisfy the inequality")
        return [(c, d)]

    # r represents the distance from x0

    r = math.sqrt(rhs)

    # The critical boundary points where the inequality becomes equality
    p1 = x0 - r
    p2 = x0 + r

    # We will store the admissible intervals here
    intervals = []

    # Left interval: p <= p1
    # intersected with the domain [c,d]
    left_end = min(p1, d)
    if c <= left_end:
        intervals.append((c, left_end))

    # Right interval: p >= p2
    # intersected with the domain [c,d]
    right_start = max(p2, c)
    if right_start <= d:
        intervals.append((right_start, d))


    return intervals
```

```python
# Source
x_min, x_max = -1, 1
y_min, y_max = x_min * 1, x_max * 1

# Discretizing
N = 100
dx = (x_max - x_min) / (N - 1)

x = np.linspace(x_min, x_max, N)
src = np.exp(-x**2 / 0.1)

y = np.linspace(y_min, y_max, N)
output = src * 1

# Refractive index
n = 1.5

beta = 5
```

```python
def compute_C_matrix(x, m, n, beta):
    # Ensure inputs are at least 2D (N, D)
    x = np.atleast_2d(x).T if x.ndim == 1 else x
    m = np.atleast_2d(m).T if m.ndim == 1 else m

    N, D = x.shape  # D will be 1 for your script
    gamma = (n**2) / (n**2 - 1)

    diff = x - m
    sq_dist = np.sum(diff**2, axis=1)

    term_inside_sqrt = beta**2 - (n**2 - 1) * sq_dist
    term_inside_sqrt = np.maximum(term_inside_sqrt, 1e-10)
    c_tilde = -(n / (n**2 - 1)) * np.sqrt(term_inside_sqrt)

    # Initialize C array (N, D, D)
    C = np.zeros((N, D, D))

    for i in range(N):
        xi_yi = diff[i].reshape(D, 1)
        outer_prod = np.dot(xi_yi, xi_yi.T)

        # Matrix C calculation based on the paper's formula [cite: 541]
        C[i] = (gamma / c_tilde[i]) * np.eye(D) + \
               (gamma**2 / (c_tilde[i]**3)) * outer_prod

    return C.squeeze() # Returns scalar C for 1D, or (N, 2, 2) for 2D
```

```python
C = compute_C_matrix(x, m, n, beta)
```

```python
def Ji(m, P, C = C, src = src, dx = dx):
  # Just numerically finds the derivative
  # Also 1D so kinda chill
  dm_dx = np.gradient(m, dx)

  error = (C * dm_dx) - P

  # Rieman sum basically
  return 0.5 * np.sum(src * (error ** 2)) * dx
```

```python
def Jb(m, b):
  # We have a surface integral and I think here I could just use the endpoints due to a 1D source
  error_start = (m[0] - b[0]) ** 2
  error_end = (m[-1] - b[-1]) ** 2

  return .5 * (error_start + error_end)
```

```python
def J(m, P, b, alpha, C = C, src = src, dx = dx):
  ji = Ji(m, P, C, src, dx)
  jb = Jb(m, b)

  return alpha * ji + (1 - alpha) * jb
```

```python
def min_P(m, x = x, src = src, output = output, y = y, C = C):
    # We dicretized it but the output of m is a specific value which we got to interpolate
    g_at_m = np.interp(m, y, output)

    # We don't need to solve the quartic using the langrangian func since in 1D P is just a 1 by 1 scalar not a 2 by 2 matrix
    # Normally would be det(P) = f/g det(C)
    # epsilon to avoid division by zero
    P = (src / (g_at_m + 1e-5)) * C
    return P
```

```python
def main_step(P, dx, N):
    # np.gradient can be finicky. Let's ensure we get a 1D array.
    grad_result = np.gradient(P, dx)

    # If grad_result is a list/tuple, the first element is the actual gradient array
    if isinstance(grad_result, (list, tuple)):
        b = grad_result[0]
    else:
        b = grad_result

    # Force b to be 1D and length N
    b = b.flatten()

    # Construct the Laplacian Matrix A
    main_diag = -2 * np.ones(N)
    off_diag = np.ones(N - 1)
    A = diags([off_diag, main_diag, off_diag], [-1, 0, 1]).tocsr()

    # Boundary Conditions
    # We fix the first point of the lens height to 0
    A[0, :] = 0
    A[0, 0] = 1
    b[0] = 0

    # Solve the system (A * 1/dx^2)u = b
    # A is the second derivative operator
    u = spsolve(A / (dx**2), b)

    # The mapping m is the first derivative of the surface u
    m_new = np.gradient(u, dx)

    return u, m_new
```

```python
import numpy as np
from scipy.sparse import diags
from scipy.sparse.linalg import spsolve

# 1. Setup based on your variables
gamma = n**2 / (n**2 - 1)
m = np.linspace(y_min, y_max, N) # Initial guess for mapping
dt = 0.2  # Relaxation factor (prevents overflow/instability)

def get_C_and_P(x, m, src, output, n, beta):
    # Calculate non-quadratic term: beta^2 - (n^2 - 1)|x - m|^2
    # We clip to 1e-8 to prevent imaginary numbers which cause overflows
    arg = beta**2 - (n**2 - 1) * (x - m)**2
    arg = np.maximum(arg, 1e-8)

    c_tilde = -(n / (n**2 - 1)) * np.sqrt(arg)

    # Paper eq for C (1D version)
    C = (gamma / c_tilde) + (gamma**2 * (x - m)**2) / (c_tilde**3)

    # Energy conservation: P = (Source / Output) * C
    g_at_m = np.interp(m, np.linspace(y_min, y_max, N), output)
    P = (src / (g_at_m + 1e-5)) * C
    return C, P

# 2. Iterative ELS Loop
for i in range(100):
    C, P = get_C_and_P(x, m, src, output, n, beta)

    # The paper minimizes J = 0.5 * ||C*m' - P||^2
    # The Euler-Lagrange eq is: (C^2 * m')' = (C * P)'

    # Constructing the weighted Laplacian A
    # We use a helper variable W = C^2 to avoid overflow in the matrix directly
    W = np.clip(C**2, 1e-5, 1e5)

    # Finite difference for (W * m')'
    main_diag = -2 * W
    off_diag = W[:-1]
    A = diags([off_diag, main_diag, off_diag], [-1, 0, 1], shape=(N, N)).tocsr()

    # Boundary Conditions: Substitution to avoid MatrixRankWarning
    # Fix mapping at boundaries to the target domain [y_min, y_max]
    rhs = np.gradient(C * P, dx) * (dx**2)
    rhs[0] = y_min
    rhs[-1] = y_max
    A[0, :] = 0; A[0, 0] = 1
    A[-1, :] = 0; A[-1, -1] = 1

    # Solve for target mapping m_hat
    try:
        m_hat = spsolve(A, rhs)
        # Relaxation step: prevents the mapping from jumping into unphysical regions
        m_new = m + dt * (m_hat - m)

        diff = np.linalg.norm(m_new - m)
        m = m_new

        if diff < 1e-6:
            print(f"Converged at iter {i}")
            break
    except:
        print("Numerical instability - reducing step size")
        dt *= 0.5

# 3. Final Surface Reconstruction (Stage 2 of the paper)
# Using the Law of Refraction: grad(u) = grad_x c(x, m)
arg_final = np.maximum(beta**2 - (n**2 - 1) * (x - m)**2, 1e-8)
c_tilde_final = -(n / (n**2 - 1)) * np.sqrt(arg_final)
du_dx = -(gamma / c_tilde_final) * (x - m)

# Integrate the gradient to find the lens surface profile
u = np.cumsum(du_dx) * dx
```

```python
import matplotlib.pyplot as plt

# Using the u and x from the previous step
plt.figure(figsize=(10, 6))
plt.plot(x, u, label='Lens Surface $u(x)$', color='blue', linewidth=2)
plt.title('Reconstructed Freeform Lens Surface (1D Profile)')
plt.xlabel('Source Coordinate $x$')
plt.ylabel('Surface Height $u$')
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.show()

# To see how well the energy was redistributed, plot the mapping m
plt.figure(figsize=(10, 6))
plt.plot(x, m, label='Mapping $m(x)$', color='red')
plt.title('Light Ray Mapping: Source to Target')
plt.xlabel('Source Position')
plt.ylabel('Target Position')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# 1. Calculate the second surface u2 using the relation from the paper
# c(x, y) = u1(x) + u2(y) => u2(y) = c(x, m(x)) - u1(x)
arg_rays = np.maximum(beta**2 - (n**2 - 1) * (x - m)**2, 1e-8)
c_val = -(n / (n**2 - 1)) * np.sqrt(arg_rays) # The variable part of cost c
u2 = c_val - u # u is the u1 we calculated earlier

# 2. Setup Plot
plt.figure(figsize=(12, 8))

# Plot Lens Surfaces
plt.plot(x, u, 'b-', linewidth=2, label='Lens Surface $\mathcal{L}_1$ (Entry)')
plt.plot(m, u2, 'r-', linewidth=2, label='Lens Surface $\mathcal{L}_2$ (Exit)')

# 3. Trace Rays
# We'll pick every 5th ray for clarity
for i in range(0, N, 5):
    # Segment 1: Source plane (z=constant) to L1
    z_src = u.min() - 0.5
    plt.plot([x[i], x[i]], [z_src, u[i]], 'g--', alpha=0.4)

    # Segment 2: Through the lens (L1 to L2)
    plt.plot([x[i], m[i]], [u[i], u2[i]], 'k-', alpha=0.6)

    # Segment 3: L2 to Target plane
    z_target = u2.max() + 0.5
    plt.plot([m[i], m[i]], [u2[i], z_target], 'g--', alpha=0.4)

# Plot Source and Target lines
plt.axhline(y=z_src, color='gray', linestyle=':', label='Source Plane')
plt.axhline(y=z_target, color='orange', linestyle=':', label='Target Plane')

plt.title('Raytrace: Parallel Source to Target Distribution')
plt.xlabel('Lateral Position')
plt.ylabel('Optical Axis (z)')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.grid(True, which='both', linestyle='--', alpha=0.5)
plt.axis('equal') # Important to see real refraction angles
plt.tight_layout()
plt.show()
```
