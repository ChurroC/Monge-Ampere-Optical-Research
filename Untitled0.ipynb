## Defined Data

```python
#define intervals in a domain [a,b]
a, b = 0.0, 1.0 #allows for the entire interval
N = 500 #grid spacing
x = np.linspace(a, b, N)

#Source and target fucntions
def f(x):
  return 1.0 + 0*x  #a = 0 it is uniform

def g(y):
  return 2.0*y +0.1 #example target function that is non-uniform

#For two lenses we must integrate to the same mass
#normalization of the same mass
Fx = np.trapezoid(f(x), x)

#this will build a grid for the target density
y = np.linspace(0.0, 1.0, N)
Gy = np.trapezoid(g(y), y)

#normalize g function so that the masses can match
def g_norm(y):
  return g(y) * Fx / Gy
```
